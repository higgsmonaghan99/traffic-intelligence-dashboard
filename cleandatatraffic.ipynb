```python
import zipfile
import os

zip_path = r"C:\Users\husnu\Downloads\drive-download-20260406T221849Z-3-001.zip"
extract_path = r"C:\Users\husnu\Downloads\panynj_extracted"

os.makedirs(extract_path, exist_ok=True)

with zipfile.ZipFile(zip_path, "r") as zip_ref:
    zip_ref.extractall(extract_path)

pdf_files = []
for root, dirs, files in os.walk(extract_path):
    for file in files:
        if file.lower().endswith(".pdf"):
            pdf_files.append(os.path.join(root, file))

pdf_files = sorted(pdf_files)

print("PDF sayısı:", len(pdf_files))
for pdf in pdf_files[:16]:
    print(pdf)
```

```python
import pandas as pd
import pdfplumber
import os
import re

all_rows = []

for pdf_path in pdf_files:
    file_name = os.path.basename(pdf_path)
    year_match = re.search(r"(20\d{2}|201\d)", file_name)
    source_year = year_match.group(1) if year_match else None

    print("İşleniyor:", file_name)

    try:
        with pdfplumber.open(pdf_path) as pdf:
            for page_num, page in enumerate(pdf.pages, start=1):
                tables = page.extract_tables()

                if not tables:
                    continue

                for table_no, table in enumerate(tables, start=1):
                    for row_no, row in enumerate(table, start=1):
                        all_rows.append({
                            "source_file": file_name,
                            "source_year": source_year,
                            "page": page_num,
                            "table_no": table_no,
                            "row_no": row_no,
                            "row_data": row
                        })

    except Exception as e:
        print("HATA:", file_name, "->", e)

df_raw = pd.DataFrame(all_rows)
print(df_raw.shape)
df_raw.head()
```

```python
output_excel = r"C:\Users\husnu\Downloads\panynj_pdf_raw.xlsx"
df_raw.to_excel(output_excel, index=False)
print("Kaydedildi:", output_excel)
```

```python
rows_expanded = pd.DataFrame(df_raw["row_data"].tolist())
df_expanded = pd.concat(
    [df_raw.drop(columns=["row_data"]).reset_index(drop=True), rows_expanded],
    axis=1
)

print(df_expanded.shape)
df_expanded.head(20)
```

```python
df = df_expanded.copy()

df["text"] = df[0].astype(str).str.strip()
df = df[["source_year", "text"]]

df.head()
```

```python
facilities = [
    "George Washington Bridge",
    "Lincoln Tunnel",
    "Holland Tunnel",
    "Goethals Bridge",
    "Outerbridge Crossing",
    "Bayonne Bridge"
]

current_facility = None
parsed_rows = []

for _, row in df.iterrows():
    text = row["text"]
    year = row["source_year"]

    # facility yakala
    if text in facilities:
        current_facility = text
        continue

    # veri satırı mı?
    if "Automobiles" in text and current_facility:
        parsed_rows.append({
            "year": year,
            "facility": current_facility,
            "raw_values": text
        })

parsed_df = pd.DataFrame(parsed_rows)

parsed_df.head()
```

```python
import re

def extract_numbers(text):
    nums = re.findall(r"\d{1,3}(?:,\d{3})+", text)
    return [int(n.replace(",", "")) for n in nums]

parsed_df["values"] = parsed_df["raw_values"].apply(extract_numbers)

parsed_df.head()
```

```python
months = [
    "Jan","Feb","Mar","Apr","May","Jun",
    "Jul","Aug","Sep","Oct","Nov","Dec"
]
```

```python
final_rows = []

for _, row in parsed_df.iterrows():
    year = row["year"]
    facility = row["facility"]
    values = row["values"]

    for i, val in enumerate(values):
        if i < 12:
            final_rows.append({
                "year": year,
                "facility": facility,
                "month": months[i],
                "traffic": val
            })

final_df = pd.DataFrame(final_rows)

final_df.head()
```

```python
final_df["traffic"] = final_df["traffic"].apply(lambda x: f"{x:,}")
```
